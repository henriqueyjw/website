```python
import pandas as pd

df = pd.read_stata('dados_estrategia_1.dta')
df = df[df.prazo >0]
df.head()              
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

sf = sm.SurvfuncRight(df["prazo"], df["status"])
sf.summary().head()
```

```python
mod = smf.phreg("prazo ~ 0 + sexo_num + C(escolaridade_num) + C(area_governo_num) + C(estado_civil_num)", df, status=df["status"], ties="efron")
rslt = mod.fit()
print(rslt.summary())
```

```python
import pandas as pd

df = pd.read_stata('logit_henrique.dta')
df.head()
```

```python
mod = smf.logit("logit_10years ~ 0 + sexo_num + C(raça) + C(escolaridade_num) + C(area_governo_num) + C(estado_civil_num)", df)
rslt = mod.fit()
margins = rslt.get_margeff()
print(margins.summary())
```

```python
df = pd.read_csv('student-mat.csv', sep = ';')
df.head()
```

```python
df = pd.get_dummies(df, drop_first = True, dtype=int)
df.head()
```

```python
mod = smf.gls("G1 ~ C(age) + Medu + Fedu + traveltime + studytime + freetime + sex_M + Mjob_teacher + Fjob_teacher + internet_yes + famrel", df)
rslt = mod.fit()
print(rslt.summary())
```

```python
import pandas as pd

df = pd.read_csv('heart_failure_clinical_records_dataset.csv')
df.head()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

sf = sm.SurvfuncRight(df["time"], df["DEATH_EVENT"])
sf.summary().head()
```

```python
mod = smf.phreg("time ~ 0 + age + anaemia +creatinine_phosphokinase + diabetes + ejection_fraction + high_blood_pressure + platelets + serum_creatinine + serum_sodium + sex + smoking", df, status=df["DEATH_EVENT"], ties="efron")
rslt = mod.fit()
print(rslt.summary())
```
