```python
import pymupdf as fitz
import easyocr

pdffile = "ocr_text.pdf"
doc = fitz.open(pdffile)
zoom = 4
mat = fitz.Matrix(zoom, zoom)
count = 0
# Count variable is to get the number of pages in the pdf
for p in doc:
    count += 1
for i in range(count):
    val = f"image_{i+1}.png"
    page = doc.load_page(i)
    pix = page.get_pixmap(matrix=mat)
    pix.save(val)
doc.close()

reader = easyocr.Reader(['pt'])
result = reader.readtext("image_1.png", detail=0)
result
```

```python
from PyPDF2 import PdfReader

# Specify the path to your PDF file
pdf_path = "normal_text.pdf"

try:
    # Create a PdfReader object
    reader = PdfReader(pdf_path)

    # Get the total number of pages
    num_pages = len(reader.pages)
    print(f"Number of pages in the PDF: {num_pages}")

    # Extract text from a specific page (e.g., the first page)
    page = reader.pages[0]  # Access the first page (index 0)
    text = page.extract_text()
    print("\nText from the first page:")
    print(text)

    # You can loop through all pages to extract all text
    all_text = ""
    for page_num in range(num_pages):
        page = reader.pages[page_num]
        all_text += page.extract_text() + "\n"

    print("\nFull text from the PDF:")
    print(all_text)

except FileNotFoundError:
    print(f"Error: The file '{pdf_path}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import os
import pandas as pd

name_report = 'titanic3.xls'
df = pd.read_excel(name_report)
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['age']).isnull().sum()
```

```python
data = df.drop(columns = ['name','home.dest','boat','body','cabin','ticket'])
data.head()
```

```python
data = pd.get_dummies(data, drop_first = True, dtype=int)
data.head()
# data.isnull().sum()
```

```python
data.to_excel('titanic_limpo.xlsx', index=False)
# data.to_csv('titanic_limpo.csv', index=False)
# data.to_json('titanic_limpo.json', index=False)
```
