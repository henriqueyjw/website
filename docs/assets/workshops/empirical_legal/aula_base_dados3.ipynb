```python
import pandas as pd
import os

df = pd.DataFrame()
for arquivo in os.listdir('aula4'):
    if arquivo.endswith('.csv'):
        df = pd.concat([df, pd.read_csv(f'aula4/{arquivo}')], ignore_index=True)
df.head()
```

```python
import re

termos = {
    'requerente': r'(?<=requerente:)(.*)(?=requerido:)',
    'requerido': r'(?<=requerido:)(.*)(?=juiz\(a\))',
}

req_te = list()
req_do = list()

for _,row in df.iterrows():
    if type(row.decisao) == str:
        for u,v in termos.items():
            regex = re.search(v, row.decisao, re.I|re.S)
            if regex:
                if u == 'requerente':
                    req_te.append(regex.group().strip())
                elif u == 'requerido':
                    req_do.append(regex.group().strip())
            else:
                if u == 'requerente':
                    req_te.append(None)
                elif u == 'requerido':
                    req_do.append(None)
    else:
        req_te.append(None)
        req_do.append(None)
        
df['requerente'] = req_te
df['requerido'] = req_do
df.head()
```

```python
import json
import pandas as pd
import os

df = pd.DataFrame()
for arquivo in os.listdir('dados_exemplo'):
    if not arquivo.endswith('edital.txt'):
        with open(f'dados_exemplo/{arquivo}', 'r+') as f:
            data = json.load(f)
        try:
            df = pd.concat([df, pd.json_normalize(data, max_level=1)], ignore_index=True)
        except:
            pass
df.head()
df.to_csv('compras_publicas.csv', index=False)
```

```python
import json
import pandas as pd
import os

df = pd.DataFrame()
texto = list()
processo = list()
for arquivo in os.listdir('dados_exemplo'):
    if arquivo.endswith('edital.txt'):
        with open(f'dados_exemplo/{arquivo}', 'r+') as f:
            data = f.read()
        texto.append(data)
        processo.append(arquivo.split('_')[0])

df['edital'] = texto
df['processo'] = processo
df.head()
df.to_csv('editais.csv', index=False)
```

```python
import re
import pandas as pd

df = pd.read_csv('editais.csv')
df.head()

termos = {
    'administrativo': r'administ',
    'judicial': r'judicial|juridico|justi[çc]a'
}

adm = list()
jud = list()

for _,row in df.iterrows():
    for u,v in termos.items():
        regex = re.search(v, row.edital, re.I|re.S)
        if regex:
            if u == 'administrativo':
                adm.append(row.edital[regex.span()[0]-20:regex.span()[1]+20])
            elif u == 'judicial':
                jud.append(row.edital[regex.span()[0]-20:regex.span()[1]+20])
        else:
            if u == 'administrativo':
                adm.append(None)
            elif u == 'judicial':
                jud.append(None)

df['administrativo'] = adm
df['judicial'] = jud
df.head()
```
