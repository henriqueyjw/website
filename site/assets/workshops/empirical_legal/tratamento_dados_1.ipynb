### Como abrir arquivos de dados?

Geralmente optamos por salvar nossos dados quantitativos (e alguns qualitativos) em planilhas ou arquivos de texto (csv). Esses arquivos possuem duas dimensões, ou seja possuem apenas linhas e colunas. 

Quando temos necessidade de olhar os mesmos indivíduos ao longo do tempo (dados longitudinais) adicionamos uma coluna que indica a data de referencia daquela informação, assim é como se tivessemos um cubo de dados, onde a terceira dimensão passa a ser o tempo.

Quando os dados são todos no mesmo instante ou a diferença temporal não importa para o que precisamos, chamamos de dados transversais e nele basicamente mantemos somente as duas dimensões originais.

### Como criar, filtrar e modificar os dados atuais?

Nem sempre a base de dados que recebemos é exatamente o que precisamos, algumas vezes precisamos modificar os dados que temos para adaptar ao que vamos utilizar. Seja criando uma coluna nova apenas com o ano, filtrando os dados que possuem uma característica específica ou indicando a existencia de um conjunto de termos expecíficos em um texto, as ferramentas abaixo vão ser uteis para a construção desse processo.

```python
import pandas as pd

df = pd.read_csv('base_dados/tratamento_dados/votacao_candidato_munzona_2022_BRASIL.csv')
df.columns
```

Caso o estudo que estamos fazendo seja focado apenas nas votações para cargos estaduais, temos que olhar colunas que ajudam a filtrar os dados e remover as informações que não temos interesse. Isso é muito importante quando estamos lidando com dados muito grandes, pois reduz a quantidade de memória usada no computador, permitindo adicionar outras informações ou dados sem afetar a eficiencia do computador.

```python
print(df.DS_ELEICAO.unique())
print(df.DS_CARGO.unique())
```

```python
df = df[df.DS_CARGO == 'Senador']
df.head()
```

```python
for _,row in df.iterrows():
    print(row)
    break
```

```python
dummy_eleito = list()

for _,row in df.iterrows():

    if row.DS_SIT_TOT_TURNO  == 'ELEITO':
        dummy_eleito.append(1)
    elif row.DS_SIT_TOT_TURNO == 'NÃO ELEITO':
        dummy_eleito.append(0)
    else:
        dummy_eleito.append(None)

df['dummy_eleito'] = dummy_eleito
df.head()
```

### Como adicionar dados na base que temos atualmente?

Referencias de como juntar base de dados https://pandas.pydata.org/docs/user_guide/merging.html

### Como criar variáveis a partir dos textos que temos?

Até agora, aprendemos a lidar com transformações simples e dados numéricos. Mas como fazemos quando temos dados de texto?
Para esse tipo de dado, podemos começar usando Expressões Regulares (Regex). Esse é o método mais simples de buscar informações a partir de certos padrões, sem requerer muita 'tecnologia'.

Para testar os padrões que queremos ver, podemos usar o site https://regex101.com/

```python
import pandas as pd

df = pd.read_csv('catedra/cade_clinicas_2.csv')

colunas_sem_dados = [
                    'palavras', 'relator', 'score', 'nucluster', 'sumario',
                    'representante_legal', 'descricao_assunto', 'quorum',
                    'tags', 'entidade', 'numero_acordao', 'interessados',
                    'numero_ata', 'data_publicacao','documentos_citados',
                    'descricao_protocolo', 'dispositivo_conclusao',
                    'representante_mp', 'selecionado', 'qtd_acesso',
                    'soundex', 'clicks', 'descricao_titulo_documento',
                    'vltfidf'
                    ]
df.drop(columns=colunas_sem_dados, inplace=True)
df.dropna(subset='corpo_texto_formatado', inplace=True)
df.columns
```
