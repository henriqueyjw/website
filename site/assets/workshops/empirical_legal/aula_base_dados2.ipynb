```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vect = TfidfVectorizer(analyzer='word', token_pattern=r'\w{1,}', max_features=5000)
docs = [doc for doc in df.decisao if isinstance(doc, str)]
tfidf_vect.fit(docs)
X =  tfidf_vect.transform(docs)
pd.DataFrame(X.toarray(), columns=tfidf_vect.get_feature_names_out()).head()
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer
from sklearn.feature_extraction.text import TfidfVectorizer

docs = [doc for doc in df.decisao if isinstance(doc, str)]
clean_text = list()
stop = stopwords.words('portuguese')
snowball_stemmer = SnowballStemmer("portuguese") # Specify language for SnowballStemmer
for text in docs:
    words = word_tokenize(text)
    separator = ' '
    filtered_words = [word.lower() for word in words if word not in stop and word.isalpha()]
    stemmed_words = [snowball_stemmer.stem(word) for word in filtered_words if len(word) > 2]
    clean_text.append(separator.join(stemmed_words))
tfidf_vect = TfidfVectorizer(analyzer='word', token_pattern=r'\w{1,}', max_features=5000)
tfidf_vect.fit(clean_text)
X =  tfidf_vect.transform(clean_text)
pd.DataFrame(X.toarray(), columns=tfidf_vect.get_feature_names_out()).head()
```
